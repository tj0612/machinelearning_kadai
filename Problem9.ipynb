```python
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
import math
```

```python
np.random.seed(10)
n=40
omega=np.random.randn(1)
noise = 0.8 * np.random.randn(n)

x1=np.random.randn(n,2)
y1=(2*(omega*x1[:,0]+x1[:,1]+noise>0)-1).T
```

```python
plt.scatter(x1[:,0],x1[:,1],c=y1)
plt.legend()
plt.show()
```

```python
n=100
x2=3*np.random.rand(n,2)-0.5
radius=[x2[:,0]**2+x2[:,1]**2]
y2=(np.array(list(map(lambda x:x>(0.7+0.1 * np.random.randn(n)),radius))[0]) & (np.array(list(map(lambda x:x <(2.2+0.1*np.random.randn(n)),radius)))))[0]
```

```python
plt.scatter(x2[:,0],x2[:,1],c=y2)
plt.legend()
plt.show()
```

```python
J=lambda x,y,w,l: sum(list(map(lambda a,b: math.log(1+math.exp(-b*np.dot(w,a))),x,y))) + l * np.dot(w,w)
dJ=lambda x,y,w,l: sum(list(map(lambda a,b:(-b*a)*(math.exp(-b*np.dot(w,a)))*(1/(1+math.exp(-b*np.dot(w,a)))),x,y))) + 2*l * w
ddJ=lambda x,y,w,l: sum(list(map(lambda a,b: (-b*a)*(-b*a)*(math.exp(-b*np.dot(w,a)))*(1/(1+math.exp(-b*np.dot(w,a)))) + 
                                                                                    (((-b*a)*(math.exp(-b*np.dot(w,a))))**2)*(-1/(1+math.exp(-b*np.dot(w,a)))**2)+2*l,x,y)))
```

```python
def bsg_update(x,y,w,alpha,l):
    return w-alpha*dJ(x,y,w,l)
```

```python
def logreg_bsg(x,y,alpha,l=0.1,iteration=50):
    J_list=list()
    plt.close()
    w=np.zeros(x.shape[1])
    ite=0
    for i in range(iteration):
        ite=i
        w_old=w
        w=bsg_update(x,y,w,alpha,l)
        J_list.append(J(x,y,w,l))
        if np.allclose(w,w_old):
            print("stop")
            break
        #print(J(x,y,w,l))
        #print(w)
    print(J(x,y,w,l))
    print(w)
    plt.plot([j for j in range(ite+1)],J_list,linewidth=2,marker="o",label="bsg")
    plt.legend()
    plt.show()
```

```python
def logreg_newton(x,y,l=0.1,iteration=50):
    w=np.array([0,0])
    ite=0
    J_list=list()
    for i in range(iteration):
        ite=i
        w_old=w
        w=w - dJ(x,y,w,l)/ddJ(x,y,w,l)
        J_list.append(J(x,y,w,l))
        if np.allclose(w,w_old):
            print("stop")
            break
    print(J(x,y,w,l))
    print(w)
    plt.plot([j for j in range(ite+1)],J_list,linewidth=2,marker="o",label="newton")
    plt.legend()
    plt.show()
```

```python
logreg_bsg(x1,y1,0.1,iteration=10000000000)
```

```python
logreg_newton(x1,y1,iteration=100000000000000)
```

```python
logreg_bsg(x2,y2,0.1,iteration=10000000000)
```
